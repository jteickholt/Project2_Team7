### Bring in the population at city level

```python
# define path to the city level data

population_file=os.path.join('..','data','census_population.csv')

# read in the city level data

population=pd.read_csv(population_file)
```

```python
# convert to dataframe, keep only fields of interest and assign better names 

population_df=pd.DataFrame(population)

population_df=population_df[['GEO_ID', 'NAME','S0101_C01_001E']]

population_df=population_df.rename(columns={'NAME': 'City', 'S0101_C01_001E' : 'Population'})

population_df.head()
# population_df.describe()

```

### Bring in the population at the state level

```python
# define path to the city level data

population_state_file=os.path.join('..','data','census_population_state.csv')

# read in the city level data

population_state=pd.read_csv(population_state_file)
```

```python
# convert to dataframe, keep only fields of interest and assign better names 

population_state_df=pd.DataFrame(population_state)

population_state_df=population_state_df[['GEO_ID', 'NAME','S0101_C01_001E']]

population_state_df=population_state_df.rename(columns={'NAME': 'State', 'S0101_C01_001E' : 'Population'})

population_state_df.head()
```

### Pull out state code

```python
city = population_df["City"].str.split(",", expand=True)

city2 = city[0]


population_df["City2"] = city2

state = population_df["City"].str.split(",", expand=True)

state2=state[1]

state3=state2.str.slice(0,3,1)

population_df["State"] = state3

population_df.head()

```

### Merge the city level data

```python
city_df=pd.merge(pd.merge(pd.merge(commute_df, income_df,on="GEO_ID", suffixes=("","_y")), unemployment_df, on="GEO_ID"), population_df, on="GEO_ID")

city_df=city_df.drop(columns=["City_x","City_y"])
city_df = city_df.rename(columns={"City": "City_Code", "City2": "City"})

city_df["City_State"]=city_df["City"] + "," + city_df["State"]
city_df.head()
```

```python
state_df=pd.merge(pd.merge(pd.merge(commute_state_df, income_state_df,on="GEO_ID", suffixes=("","_y")), unemployment_state_df, on="GEO_ID"), population_state_df, on="GEO_ID")
state_df=state_df.drop(columns=["State_x","State_y"])
state_df.head()
```

### Bring in csv files that Luis created for diversity and education data from census

```python
# define path to the city level data

divedu_file=os.path.join('.','city_final.csv')

# read in the city level data

div_edu=pd.read_csv(divedu_file)
```

```python
div_edu_df=pd.DataFrame(div_edu)


# drop the first row as it was the second header row in the original file
div_edu_df = div_edu_df.drop([0])

# drop the Area Name as it will already be in the file that we will merge this with
# also drop this other columns which appears to be a duplicate of Less than 9th Grade
div_edu_df = div_edu_df.drop(columns=["Area Name_x", "Less than 9th Grade.1"])

div_edu_df.head()
```

```python
# define path to the state level data

divedu_state_file=os.path.join('.','state_final.csv')

# read in the state level data

div_edu_state=pd.read_csv(divedu_state_file)
```

```python
div_edu_state_df=pd.DataFrame(div_edu_state)


# drop the first row as it was the second header row in the original file
div_edu_state_df = div_edu_state_df.drop([0])

# drop the Area Name as it will already be in the file that we will merge this with
div_edu_state_df = div_edu_state_df.drop(columns=["Area Name_x"])

div_edu_state_df.head()
```

#### Merge the files into the master city and state file

```python
city_final_df=pd.merge(city_df, div_edu_df, on="GEO_ID")

city_final_df.head()
```

```python
state_final_df=pd.merge(state_df, div_edu_state_df, on="GEO_ID")

state_final_df.head()
```
